#### 1. babbling 안에 word가 포함되어 있나?

```python
word = ["aya", "ye", "woo", "ma"]
babbling = ["aya", "yee", "u", "maa", "wyeoo"]
for i in range(len(babbling)):
    contain = 0
    for j in range(len(word)):
        if word[j] in babbling[i]:
            contain += 1
            print("i : ", babbling[i] , " j : ", word[j])
        else:
            pass
```

#### 2. word의 조합이 babbling 전체가 되는가?
: replace("특정 문자열", "대체 문자열")
: 단, 중간에 존재할 경우도 지워버린다.

```python
word = ["aya", "ye", "woo", "ma"]
babbling = ["aya", "yee", "u", "maa", "wyeoo"]
for i in range(len(babbling)):
    contain = 0
    for j in range(len(word)):
        if word[j] in babbling[i]:
            contain += 1
            print("i : ", babbling[i] , " j : ", word[j])
            babbling[i] = babbling[i].replace(word[j],"")
            print("babbling [",i,"] : ", babbling[i])
        else:
            pass
```

#### 3. 삭제할 word의 위치가 0이 아니라면, 삭제 되지 않도록!

```python
contain = 0
word = ["aya", "ye", "woo", "ma"]
babbling = ["ayaye", "uuuma", "ye", "yemawoo", "ayaa"]
for i in range(len(babbling)):
    for j in range(len(word)):
        if word[j] in babbling[i]:
            if babbling[i].find(word[j]) == 0:
                babbling[i] = babbling[i].replace(word[j],"")
                print("babbling [",i,"] : ", babbling[i])
            
                if(babbling[i] == ""):
                    contain += 1
        else:
            pass
print(contain)
```

#### 4. word의 길이만큼만 for문을 돌면 누락 건수가 생김!

```python
answer = 0
word = ["aya", "ye", "woo", "ma"]
babbling = ["mawooye", "woo", "u", "maaya", "wyeoo"]
for i in range(len(babbling)):
    j = 0
    while(j != len(word)):
        babbling[i] = babbling[i].replace(word[j]," ")
        print("babbling [",i,"] : ", babbling[i])
        j += 1
#   공백을 삽입했다가 비교후에 공백 삭제 -> 단어가 삭제된 후에 조합되는 문제를 방지
    babbling[i] = babbling[i].replace(" ", "")
    if(babbling[i] == ""):
        answer += 1
        print(i, "번째 : ", answer)
print("옹알이 개수 : ", answer)
```
