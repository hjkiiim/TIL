## 1. 나만의 지도 그리기
### 내가 사는 지역의 박물관/미술관은 어디 있을까❓❓

```python
import folium
import pandas as pd
df = pd.read_excel("./file/museum.xls")
mymap = folium.Map(location=[37.3082,127.9295],zoom_start=10)
df = df.dropna(subset=['위도','경도','소재지지번주소'])
df = df.query("소재지지번주소.str.contains('강원')", engine='python')
for i in range(len(df)):
    folium.Marker([float(df['위도'][i:i+1]), float(df['경도'][i:i+1])], tooltip=df['시설명'][i:i+1], icon=folium.Icon(color="red")).add_to(mymap)

mymap
```

## 2. 나만의 차트 그리기
### 내가 사는 지역의 강수량 비교하기 (강원도 2019년 강수량 자료)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_excel("./file/rain_20191231.xls")
plt.rc('font', family='NanumGothic')
plt.plot(df['시군별'],df['1월'], 'bo', label="1월", marker="*", color="skyblue")
plt.plot(df['시군별'],df['3월'], 'bo', label="3월", marker="v", color="gold")
plt.plot(df['시군별'],df['5월'], 'bo', label="5월", marker=".", color="green")
plt.plot(df['시군별'],df['7월'], '--', label="7월", marker="$♥$", color="red")
plt.plot(df['시군별'],df['9월'], ':', label="9월", marker="+", color="orange")
plt.plot(df['시군별'],df['12월'], 'bo', label="12월", marker="p", color="silver")
plt.title("강원도 강수량")
plt.legend(loc=(1.1,0.65))
plt.grid(True, axis='y')
plt.xticks(np.arange(0,18,2))
plt.show()
```
